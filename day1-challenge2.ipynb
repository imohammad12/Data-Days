```python
data_url = "/Users/mohammad/Documents/Internship-IAI/Data-days-stuff/divar_posts_dataset.csv"
data = pd.read_csv(data_url)
```

```python
def plot(columns, filters = [], types='single', sold = False):
    if types == 'single':
        y = data[columns[0]].value_counts()
        plt.bar(y.index, y.values)
        plt.xticks(rotation = 90)
        return plt.show()
    if types == 'double':
        if sold == True:
            d = data.groupby(columns[0])['archive_by_user'].sum()
            plt.bar(d.index, d.values)
            plt.xticks(rotation=90)
            return plt.show()

        else:
            d = data[data[columns[0]] == filters[0]]
            y = d[columns[1]].value_counts()
            plt.bar(y.index, y.values)
            plt.xticks(rotation = 90)
            return plt.show()
```
